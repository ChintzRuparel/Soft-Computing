## Experiment 1: MCP Model for different GATES (AND, OR, ANDNOT, XOR)

### AND Gate

```python
# Aajinkya C099, BTI C2, SC
# hardcoding AND TT and weights
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]
y = [1, 0, 0, 0]
w1 = 1
w2 = 1

yin = []
for i in range(len(x1)):
    yin.append((x1[i] * w1) + (w2 * x2[i]))

theta = yin[0]

print("Final output:")
print("x1 \t x2  y \t yin")
for x_1, x_2, y_tt, y_op in zip(x1, x2, y, yin):
    print(x_1, "\t", x_2, "\t", y_tt, "\t", y_op)
    if y_tt == 1 and y_op <= theta:
        theta = y_op

print("Threshold value calculated:", theta)
print("\n<============================>\n")
print("Final output with f(yin):")
print("x1 \t x2  y \t yin f(yin)")
for x_1, x_2, y_tt, y_op in zip(x1, x2, y, yin):
    print(x_1, "\t", x_2, "\t", y_tt, "\t",
          y_op, "\t", 1 if y_op >= theta else 0)
```

## Experiment 2: MCP Model for Advanced Gates (XOR, NAND, NOR Gate)

### XOR Gate

```python
# Aajinkya C099, BTI C2, SC
# hardcoding XOR Extended TT and weights
x1 = [1, 1, 0, 0]
x2 = [1, 0, 1, 0]


# y is 1 when both inputs are different

y = [0, 1, 1, 0]
w1 = 1
w2 = 1


def printDivider():
    print("\n<============================>\n")


def doMPModelCalculation(x1, x2, y, w1, w2, skipFinal=False, dontShowF=False):
    yin = []
    for i in range(len(x1)):
        yin.append((x1[i] * w1) + (w2 * x2[i]))

    theta = yin[0]

    print("Final output:")
    print("x1 \t x2 \t y \t yin")
    for x_1, x_2, y_tt, y_op in zip(x1, x2, y, yin):
        print(x_1, "\t", x_2, "\t", y_tt, "\t", y_op)
        if skipFinal:
            if y_tt == 1 and y_op <= theta:
                theta = y_op
        else:
            if y_tt == 1:
                theta = y_op

    printDivider()

    print("Theta calculated:", theta)

    if skipFinal:
        print(
            f"\nConclusion: We can't solve the above question using the normal with the current threshold value (Î¸: {theta})\nas it doesn't satisfy all the conditions for the XOR Gate.")
    else:
        if dontShowF:
            print("Final output with f(yin):")
            print("x1 \t x2  y \t yin f(yin)")
            for x_1, x_2, y_tt, y_op in zip(x1, x2, y, yin):
                print(x_1, "\t", x_2, "\t", y_tt, "\t",
                      y_op, "\t", 1 if y_op >= theta else 0)
            printDivider()


doMPModelCalculation(x1, x2, y, w1, w2, True)

# Calculating Z1 and Z2
printDivider()
print("Calculating with yin = (z1 * w1) + (z2 * w2)")
print("Where z1 = (x1` * x2) \t z2 = (x1 * x2`)")

printDivider()

# calculating Z1 first
print("Calculating Z1 = (x1` * x2)")
doMPModelCalculation(x1, x2, [0, 0, 1, 0], -1, 1, False, False)

# calculating Z2 first
print("Calculating Z2 = (x1 * x2`)")
doMPModelCalculation(x1, x2, [0, 1, 0, 0], 1, -1, False, False)

# calculating Z1 or Z2
print("x1 \t x2 \t y \t z1 \t z2")
for x_1, x_2, y_tt, z_1, z_2 in zip(x1, x2, y, [0, 0, 1, 0], [0, 1, 0, 0]):
    print(x_1, "\t", x_2, "\t", y_tt, "\t",
          z_1, "\t", z_2)

# final statement
printDivider()
print("Thus, yin = (z1 * w1) + (z2 * w2), where v1 = v2 = 1")
print("Therefore, we realized XOR function using MP Model")
```

### NAND Gate

```python
import numpy as np

def threshold(inputs,outputs,weights):
  yin = []
  for i in inputs:
    yin.append(sum(np.multiply(i,weights).tolist()))
  thresh = []
  print("Theta values: ", yin)
  for j in range(len(outputs)):
    if outputs[j] == 1:
      thresh.append(yin[j])
  theta = min(thresh)
  return theta

inputs_nand = [[0,0],[0,1],[1,0],[1,1]]
outputs_nand = [0,0,0,1]
weights_nand = [2,2]
nand_theta = threshold(inputs_nand,outputs_nand,weights_nand)
print("NAND Gate theta: ",nand_theta)
```

## Experiment 3: Perceptron Model (AND, OR, NOT, NAND, NOR)

### AND Perceptron

```python
# Aajinkya C099, BTI C2, SC, AND Perceptron
# hardcoding AND TT and weights
x1 = [1, 1, -1, -1]
x2 = [1, -1, 1, -1]
t = [1, -1, -1, -1]
alpha = 1


def f(yin):
    return 1 if yin > 0 else 0 if yin == 0 else -1

# this function handles each iteration
# of the epoch and goes through it and calculates
# yin and gives us True if y == t, False else wise
# params:
# i == index of iteration
# w1 and w2 =  weights 1 and 2
# b = bias weight
# targets = [] of boolean for storing y == t value for each iteration


def epochIteration(i, w1, w2, b, targets=[]):

    print(
        f"\nRunning iteration ({i + 1} / {len(t)})...\n==> x1 = {x1[i]} | x2 = {x2[i]} | t = {t[i]} | w1 = {w1} | w2 = {w2} | b = {b}")
    yin = b + (x1[i] * w1) + (x2[i] * w2)
    print(f"yin = {b} + ({x1[i]} * {w1}) + ({x2[i]} * {w2})")
    print(f"yin = {yin}")
    y = f(yin)
    print(f"Calculating, y = f(yin) = {y}")
    print(f"Since y (ie. {y}) == t (ie. {t[i]}), therefore we won't update the weight." if y ==
          t[i] else f"Since y (ie. {y}) == t (ie. {t[i]}), therefore we update the weight.")

    # calculating new weights
    w1_new = w1 + (alpha * t[i] * x1[i]) if y != t[i] else w1
    w2_new = w2 + (alpha * t[i] * x2[i]) if y != t[i] else w2
    b_new = b + (alpha * t[i]) if y != t[i] else b
    targets.append(y == t[i])
    if i == len(t) - 1:
        # get out of recursion and return
        return (targets, w1_new, w2_new, b_new)
    else:
        return epochIteration(i + 1, w1_new, w2_new, b_new, targets)

# this function runs for each epoch with len(t) iterations


def epoch():
    i = 0
    flag = True
    w1 = w2 = b = 0
    while flag:
        print(f"\n\n====== Running epoch {i + 1} ======")
        (output, w1_n, w2_n, b_n) = epochIteration(i, w1, w2, b, [])
        w1 = w1_n
        w2 = w2_n
        b = b_n
        res = all(iter == True for iter in output)
        i += 1
        if res:
            print("\n=> No change in weight in this epoch, stopping.")
            print(
                "The perceptron targets and calculated output converges for all the pattern")
            flag = False


# running epochs
epoch()
```

## Experiment 4: N-Dimensional Pattern Using Perceptron

### For "I" and "J"

```python
# Aajinkya, C099, Div C, C2,
# N Dimensional Pattern -> SC
x1 = [1, 1]
x2 = [1, 1]
x3 = [1, 1]
x4 = [-1, 1]
x5 = [1, 1]
x6 = [-1, 1]
x7 = [1, 1]
x8 = [1, -1]
x9 = [1, -1]
t = [1, -1]
alpha = 1

def f(yin):
    return 1 if yin > 0 else 0 if yin == 0 else -1

# this function handles each iteration
# of the epoch and goes through it and calculates
# yin and gives us True if y == t, False else wise
# params:
# i == index of iteration
# w1 and w2 =  weights 1 and 2
# b = bias weight
# targets = [] of boolean for storing y == t value for each iteration

def epochIteration(i, w1, b, targets=[]):
    print(
        f"\nRunning iteration ({i + 1} / {len(t)})...\n==> x1 = {x1[i]} | x2 = {x2[i]} | x3 = {x3[i]} | x4 = {x4[i]} | x5 = {x5[i]} | x6 = {x6[i]} | x7 = {x7[i]} | x8 = {x8[i]} | x9 = {x9[i]} | t = {t[i]} | w = {w1} | b = {b}")
    yin = b + (x1[i] * w1[0]) + (x2[i] * w1[1]) + (x3[i] * w1[2]) + (x4[i] * w1[3]) + \
        (x5[i] * w1[4]) + (x6[i] * w1[5]) + \
        (x7[i] * w1[6]) + (x8[i] * w1[7]) + (x9[i] * w1[8])
    print(f"yin = {b} + ({x1[i]} * {w1[0]}) + ({x2[i]} * {w1[1]}) + ({x3[i]} * {w1[2]}) + ({x4[i]} * {w1[3]}) + ({x5[i]} * {w1[4]}) + ({x6[i]} * {w1[5]}) + ({x7[i]} * {w1[6]}) + ({x8[i]} * {w1[7]}) + ({x9[i]} * {w1[8]})")
    print(f"yin = {yin}")
    y = f(yin)
    print(f"Calculating, y = f(yin) = {y}")
    print(f"Since y (ie. {y}) == t (ie. {t[i]}), therefore we won't update the weight." if y ==
          t[i] else f"Since y (ie. {y}) == t (ie. {t[i]}), therefore we update the weight.")

    # calculating new weights
    temp = []
    for j in range(9):
        w1_new = w1[j]
        if j == 0:
            w1_new = w1[j] + (alpha * t[i] * x1[i]) if y != t[i] else w1[j]
        elif j == 1:
            w1_new = w1[j] + (alpha * t[i] * x2[i]) if y != t[i] else w1[j]
        elif j == 2:
            w1_new = w1[j] + (alpha * t[i] * x3[i]) if y != t[i] else w1[j]
        elif j == 3:
            w1_new = w1[j] + (alpha * t[i] * x4[i]) if y != t[i] else w1[j]
        elif j == 4:
            w1_new = w1[j] + (alpha * t[i] * x5[i]) if y != t[i] else w1[j]
        elif j == 5:
            w1_new = w1[j] + (alpha * t[i] * x6[i]) if y != t[i] else w1[j]
        elif j == 6:
            w1_new = w1[j] + (alpha * t[i] * x7[i]) if y != t[i] else w1[j]
        elif j == 7:
            w1_new = w1[j] + (alpha * t[i] * x8[i]) if y != t[i] else w1[j]
        elif j == 8:
            w1_new = w1[j] + (alpha * t[i] * x9[i]) if y != t[i] else w1[j]

        temp.append(w1_new)

    b_new = b + (alpha * t[i]) if y != t[i] else b
    targets.append(y == t[i])
    if i == len(t) - 1:
        # get out of recursion and return
        return (targets, temp, b_new)
    else:
        return epochIteration(i + 1, temp, b_new, targets)

# this function runs for each epoch with len(t) iterations

def epoch():
    i = 0
    flag = True
    w = [0] * 9
    b = 0
    while flag:
        print(f"\n\n====== Running epoch {i + 1} ======")
        (output, w, b) = epochIteration(0, w, b, [])
        res = all(iter == True for iter in output)
        i += 1
        if res:
            print("\n=> No change in weight in this epoch, stopping.")
            print(
                "The perceptron targets and calculated output converges for all the pattern")
            flag = False

# running epochs
epoch()
```

## Experiment 5: Fuzzy Set Operation

```python
# Aajinkya, C099, BTI, C, C2
# All fuzzy set operations
A = {"x1": 0.1, "x2": 0.7, "x3": 1, "x4": 0}
B = {"x1": 0.3, "x2": 0.4, "x3": 1, "x4": 0.2}

union, intersection, complimentA, complimentB, difference, sum, product, bSum, bDiff = {
}, {}, {}, {}, {}, {}, {}, {}, {}

for a, b in zip(A, B):
    union[a] = max(A[a], B[b])
    intersection[a] = min(A[a], B[b])
    complimentA[a] = round(1 - A[a], 1)
    complimentB[b] = round(1 - B[b], 1)
    difference[a] = min(A[a], 1 - B[b])
    product[a] = round(A[a] * B[b], 1)
    sum[a] = round((A[a] + B[b]) - (A[a] * B[b]), 1)
    bSum[a] = min(1, A[a] + B[b])
    bDiff[a] = round(max(0, A[a] - B[b]), 1)

print("Union:\t\t\t", union)
print("Intersection:\t\t", intersection)
print("Compliment A:\t\t", complimentA)
print("Compliment B:\t\t", complimentB)
print("Difference:\t\t", difference)
print("Algebraic Product:\t", product)
print("Algebraic Sum:\t\t", sum)
print("Bounded Sum:\t\t", bSum)
print("Bounded Product:\t", bDiff)
```

## Experiment 6: Composite Relations (Min-Max and Max-Product)

```python
# Aajinkya, C099, BTI, C, C2
# min-max and max-product compositions
A = [
    [0.6, 0.3],
    [0.2, 0.9],
]

B = [
    [1, 0.8],
    [0.5, 0.4],
    [0.3, 0.7],
]

minMax, maxProduct = [], []

for i, a in enumerate(A):
    tempMM = []
    tempMP = []
    for j, b in enumerate(B):
        # for min max here
        tempMM.append(
            max(
                min(a[0], b[0]),
                min(a[1], b[1]),
            )
        )
        # for max product here
        tempMP.append(
            round(
                max(
                    (a[0] * b[0]),
                    (a[1] * b[1]),
                ),
                2
            )
        )
    minMax.append(tempMM)
    maxProduct.append(tempMP)


def printMatrix(matrix):
    for r in matrix:
        for c in r:
            print(format(c, '.2f'), end="\t\t")
        print()


print("<=========Min Max Composition=========>")
printMatrix(minMax)

print("\n\n<=========Max Product Composition=========>")
printMatrix(maxProduct)
```

## Experiment 9: Genetic Algorithm (Word-Matching Algorithm)

```python
import random
# utility functions


def wordSimilarity(w1, w2):
    score = 0
    for i, w in enumerate(w1):
        score = (score + 1) if w2[i] == w else score
    return score


def coinToss():
    return random.randint(0, 1)


# step 1: initial population as user input
word = input("Enter the word to be guessed: ")

# step 2: initialize population using user
population = list(input("Enter population (comma seperated): ").split(", "))

# flag for repeating the process
flag = True
while flag:
    # step 3: evaluating the population by calculating the fitness value
    # fitness value => correct letter on correct place
    print("\nEvaluating Population (using fitness value => correct letter on correct place):")
    population_fitness = []
    for pop in population:
        if len(pop) != len(word):
            print(
                f"Population word (i.e. {pop}) length should be same as the length of the word (i.e. {len(word)})")
        population_fitness.append(
            [pop, wordSimilarity(pop, word)])
        print(f"{pop}\t=>\t{wordSimilarity(pop, word)}")

    # step 4: parent selection based on fitness value => using Rank Selection
    print("\nSelecting Breeding Population (using Rank Selection):")
    sortedPopulation = sorted(
        population_fitness, key=lambda tup: tup[1], reverse=True)

    selectedParents = sortedPopulation[0:2]
    print(f"1. {selectedParents[0][0]}")
    print(f"2. {selectedParents[1][0]}")

    # step 5: creating new population using crossover => uniform crossover
    print("\nPerforming Uniform Crossover (using coin toss):")
    for i in range(len(selectedParents[0][0])):
        # swapping 2 genes of both the parents based on coin tossing
        # if coin toss = 1, we swap
        # else, we keep it as it is
        val = coinToss()
        if val == 1:
            temp1 = selectedParents[0][0][:i] + \
                selectedParents[1][0][i] + selectedParents[0][0][i+1:]
            temp2 = selectedParents[1][0][:i] + \
                selectedParents[0][0][i] + selectedParents[1][0][i+1:]
            selectedParents[0][0], selectedParents[1][0] = temp1, temp2
    print(f"1. {selectedParents[0][0]}")
    print(f"2. {selectedParents[1][0]}")

    # step 6: checking fitness of offsprings
    offspring_fitness = []
    for offspring in selectedParents:
        offspring_fitness.append(
            [offspring[0], wordSimilarity(offspring[0], word)])
        print(f"{offspring[0]}\t=>\t{wordSimilarity(offspring[0], word)}")

    offspring_fitness = sorted(
        offspring_fitness, key=lambda tup: tup[1], reverse=True)
    print("\nHighest fitness value from above offsprings:")
    print(f"{offspring_fitness[0][0]}\t=>\t{offspring_fitness[0][1]}")

    # step 7: checking for threshold of 60% fitness value (if yes, then terminate)
    print("\nChecking if this highest fitness valued offspring is above the 60% threshold:")
    if (offspring_fitness[0][1] / len(word)) >= 0.6:
        print(f"======\tOffspring gotten: {offspring_fitness[0][0]}\t======")
        flag = False
    else:
        print("Threshold condition of 60% not met, retrying...")
```
